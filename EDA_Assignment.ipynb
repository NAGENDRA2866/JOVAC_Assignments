"Task 3"

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='subscription_type', y='watch_hours', errorbar=None)
plt.title('Average Watch Hours by Subscription Type')
plt.show()
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='subscription_type', y='monthly_fee', errorbar=None)
plt.title('Average Monthly Fee by Subscription Type')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='region', y='watch_hours', errorbar=None)
plt.title('Average Watch Hours by Region')
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='region', y='monthly_fee', errorbar=None)
plt.title('Average Monthly Fee by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='device', y='watch_hours', errorbar=None)
plt.title('Average Watch Hours by Device')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='device', y='monthly_fee', errorbar=None)
plt.title('Average Monthly Fee by Device')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(data=df, x='favorite_genre', y='avg_watch_time_per_day', estimator='mean', errorbar=None)
plt.title('Average Watch Time per Day by Favorite Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head(10)
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='gender', y='churned', estimator='mean', errorbar=None)
plt.title('Churn Rate by Gender')
plt.ylabel('Churn Rate')
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='region', y='churned', estimator='mean', errorbar=None)
plt.title('Churn Rate by Region')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='subscription_type', y='churned', estimator='mean', errorbar=None)
plt.title('Churn Rate by Subscription Type')
plt.ylabel('Churn Rate')
plt.show()

plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='payment_method', y='churned', estimator='mean', errorbar=None)
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.show()
```

"Task 4"

```python
numerical_cols = ['age', 'watch_hours', 'last_login_days', 'monthly_fee',
                  'number_of_profiles', 'avg_watch_time_per_day', 'churned']

correlation = df[numerical_cols].corr()
```

```python
sns.heatmap(correlation,annot=True, linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
